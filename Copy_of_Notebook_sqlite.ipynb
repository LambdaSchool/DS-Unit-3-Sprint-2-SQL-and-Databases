```
# imports
import sqlite3
import pandas as pd
from IPython.core.magic import (register_line_magic, register_cell_magic,
                                register_line_cell_magic)
```

```
# Download sqlite db
!wget 'https://github.com/LambdaSchool/DS-Unit-3-Sprint-2-SQL-and-Databases/raw/master/module1-introduction-to-sql/rpg_db.sqlite3'
```

```
# load the sqlite database
conn = sqlite3.connect('rpg_db.sqlite3')
```

```
# Custom notebook magic commands for loading sql.
@register_line_cell_magic
def sql_df(line_query, cell_query=None, conn=conn):
  if cell_query==None:
    return pd.read_sql(line_query, conn)
  return pd.read_sql(cell_query, conn)
```

```
%sql_df SELECT * FROM charactercreator_character LIMIT 10;
```

```
%%sql_df
SELECT * FROM charactercreator_character LIMIT 10;
```

```
# list tables
%%sql_df
SELECT 
    name
FROM 
    sqlite_master 
WHERE 
    type ='table' AND 
    name NOT LIKE 'sqlite_%';
```

```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect('buddymove_holidayiq.sqlite3')
```

```python
buddy = pd.read_csv('https://raw.githubusercontent.com/donw385/DS-Unit-3-Sprint-2-SQL-and-Databases/master/module1-introduction-to-sql/buddymove_holidayiq.csv')
```

```python
buddy.head()
```

```python
buddy.to_sql(name='review',con=connection, if_exists="append", index=False)
```

```python
connection.execute("SELECT * FROM review").fetchall()
```

```python
query = 'SELECT count() FROM review;'
print ('Rows:',connection.cursor().execute(query).fetchone()[0], '\n')
```

```python
df2 = pd.read_sql("SELECT * FROM review", connection)
```

```python
import pandas as pd
import numpy as np
import sqlite3
import os
```

```python
os.listdir()
```

```python
# Read in data to df
reviews = pd.read_csv('buddymove_holidayiq.csv')
```

```python
# Rename User Id to Id so theres no syntax issues later
reviews.rename(columns={'User Id': 'Id'}, inplace=True)
```

```python
# Look at df
reviews
```

```python
# Confirm zero nulls
reviews.isnull().sum()
```

```python
# Open connection to new database 
conn = sqlite3.connect('buddymove_holidayiq.sqlite3')
```

```python
# Add reviews df as a table in db
reviews.to_sql('review', conn, index=False)
```

```python
# Confirm db is listed
os.listdir()
```

```python
# Close connection
conn.close()
```

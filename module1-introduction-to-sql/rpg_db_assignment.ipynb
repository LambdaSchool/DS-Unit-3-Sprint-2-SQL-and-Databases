```python
# How many Items does each character have? (Return first 20 rows)
for row in cursor.execute('''SELECT * FROM charactercreator_character_inventory'''):
    print(row)
```

```python
# (character_id, total number of items) for each of the first 20 characters
for row in cursor.execute(
    '''SELECT character_id, COUNT(*) 
    FROM charactercreator_character_inventory 
    GROUP BY character_id LIMIT 20'''
):
    print(row)
```

```python
# How many Weapons does each character have? (Return first 20 rows)
# Quod eveniet is armed to the teeth!
for row in cursor.execute(
    '''      SELECT character_id, COUNT(item_ptr_id) 
               FROM charactercreator_character_inventory 
    LEFT OUTER JOIN armory_weapon
                 ON charactercreator_character_inventory.item_id = armory_weapon.item_ptr_id
           GROUP BY character_id
              LIMIT 20'''
):
    print(row)
```

```python
# On average, how many Items does each Character have?
cursor.execute(
    '''SELECT AVG(CountOfItems) FROM (
    SELECT COUNT(*) AS CountOfItems
    FROM charactercreator_character_inventory 
    GROUP BY character_id
    )
    '''
).fetchone()
```

```python
# On average, how many Weapons does each character have?
# I believe this is the number of weapons per *character that has 1+ weapons*
# So I need to come up with something else
cursor.execute(
    '''SELECT AVG(CountOfWeapons) FROM ( 
    SELECT COUNT(*) AS CountOfWeapons
    FROM charactercreator_character_inventory 
    INNER JOIN armory_weapon
    ON charactercreator_character_inventory.item_id = armory_weapon.item_ptr_id
    GROUP BY character_id
    )
    '''
).fetchone()
```

```python
# It works!
cursor.execute(
    '''      SELECT AVG(CountOfWeapons) FROM (
             SELECT COUNT(item_ptr_id) AS CountOfWeapons
               FROM charactercreator_character_inventory 
    LEFT OUTER JOIN armory_weapon
                 ON charactercreator_character_inventory.item_id = armory_weapon.item_ptr_id
           GROUP BY character_id
           )
    '''
).fetchone()
```
